## Python Code

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mass = 1200      # kg
power = 80_000   # watts (150 kW)
drag_coeff = 0.29
frontal_area = 2.2
air_density = 1.225
rolling_resistance = 0.015
```

```python
dt = 0.01  # time step
time = [0] #both are list 
velocity = [0] 
acceleration = [0]
distance = [0]
```

```python
while velocity[-1] <= 27.78:  # 100 km/h in m/s
    v = velocity[-1]

    # Resistive forces
    F_drag = 0.5 * drag_coeff * frontal_area * air_density * v**2 #drag force 
    F_rr = mass * 9.81 * rolling_resistance #rolling resistance 

    # Available driving force from power (F = P / v, limit when v=0)
    F_drive = power / max(v, 1)  #engine power 

    # Net force
    F_net = F_drive - F_drag - F_rr

    # Acceleration
    a = F_net / mass

    # Euler integration
    new_v = v + a * dt
    new_t = time[-1] + dt

    velocity.append(max(new_v, 0)) #the max value between new velocity and 0 is appended to the list
    time.append(new_t)
    acceleration.append(a)
    new_s = distance[-1] + v * dt
    distance.append(new_s)
zero_to_hundred_time = time[-1]
print("0-100 km/h time:", zero_to_hundred_time , "seconds")
```

## Results

## Graphs

### Distance - Time graph

```python
plt.figure(figsize=(8,5))
plt.plot(time, distance)
plt.xlabel("Time (s)")
plt.ylabel("distance (m)")
plt.title("0–100 km/h speed Curve")
plt.grid(True)
plt.show()
```

### Velocity - Time graph

```python
plt.figure(figsize=(8,5))
plt.plot(time, velocity)
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("0–100 km/h Acceleration Curve")
plt.grid(True)
plt.show()
```

### Acceleration - Time Graph

```python
plt.figure(figsize=(8,5))
plt.plot(time, acceleration)
plt.xlabel("Time (s)")
plt.ylabel("Accelearation (m^2/s)")
plt.title("Acceleration vs time graph")
plt.grid(True)
plt.show()
```
